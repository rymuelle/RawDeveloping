```python
def random_plane(dim):

    d1 = np.random.randn(dim)
    d1 /= np.linalg.norm(d1)

    d2 = np.random.randn(dim)
    d2 -= d1 * np.dot(d1, d2)
    d2 /= np.linalg.norm(d2)

    return d1, d2
```

```python
def generate_candidates(center, d1, d2, step):

    offsets = []

    for y in [1, 0, -1]:
        for x in [-1, 0, 1]:

            offset = (
                x * step * d1 +
                y * step * d2
            )

            offsets.append(center + offset)

    return offsets
```

```python
import cv2
import optuna
import numpy as np
import random


ratings = {}
images = {}
params_store = {}

INITIAL_RATING = 1500
K = 32
```

```python
def render_image(params):

    corr_raw = raw.astype(np.float32) / 255.0

    corr_raw = pipeline(
        corr_raw,
        **params
    )

    corr_raw = np.clip(corr_raw, 0, 1)

    return (corr_raw * 255).astype(np.uint8)
```

```python
def update_elo(rA, rB, winner):

    expected_A = 1 / (1 + 10 ** ((rB - rA) / 400))
    expected_B = 1 - expected_A

    score_A = 1 if winner == "A" else 0
    score_B = 1 - score_A

    rA += K * (score_A - expected_A)
    rB += K * (score_B - expected_B)

    return rA, rB
```

```python
def compare_images(imgA, imgB):

    display = np.hstack([imgA, imgB])

    cv2.imshow(
        "A (left) vs B (right)",
        cv2.cvtColor(display, cv2.COLOR_RGB2BGR)
    )

    while True:

        key = cv2.waitKeyEx(0)

        if key in (ord("a"), ord("A")):
            return "A"

        if key in (ord("b"), ord("B")):
            return "B"
```

```python
def objective(trial):

    params = {
        "gamma": trial.suggest_float("gamma", 0.2, 1.2),
        "alpha": trial.suggest_float("alpha", 1.0, 10.0),
        "beta": trial.suggest_float("beta", 0.05, 1.0),
        "exposure_value": trial.suggest_float("exposure_value", -2, 3),
        "white": trial.suggest_float("white", 0.7, 1.5),
        "black": trial.suggest_float("black", -0.2, 0.2),
        "r": trial.suggest_float("r", 0.8, 1.2),
        "g": trial.suggest_float("g", 0.8, 1.2),
        "b": trial.suggest_float("b", 0.8, 1.2),
        "sat": trial.suggest_float("sat", 0.5, 2.0),
        "method": "logistic",
    }

    candidate_id = trial.number

    candidate_img = render_image(params)

    ratings[candidate_id] = INITIAL_RATING
    images[candidate_id] = candidate_img
    params_store[candidate_id] = params

    # first trial
    if len(images) == 1:
        return INITIAL_RATING

    opponent_id = random.choice(
        [k for k in images.keys() if k != candidate_id]
    )

    opponent_img = images[opponent_id]

    winner = compare_images(
        opponent_img,
        candidate_img
    )

    rA = ratings[opponent_id]
    rB = ratings[candidate_id]

    if winner == "A":
        rA, rB = update_elo(rA, rB, "A")
    else:
        rA, rB = update_elo(rA, rB, "B")

    ratings[opponent_id] = rA
    ratings[candidate_id] = rB

    return rB
```

```python
import cv2
import numpy as np


def find_matches(img1, img2,
                 nfeatures=10000,
                 ratio_test=0.75):
    """
    Find SIFT correspondences between two images.

    Parameters
    ----------
    img1 : ndarray
        Source image.
    img2 : ndarray
        Target image.

    Returns
    -------
    src_pts : (N,2) float32
        Coordinates in img1.
    dst_pts : (N,2) float32
        Matching coordinates in img2.
    """

    if img1.ndim == 3:
        gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    else:
        gray1 = img1

    if img2.ndim == 3:
        gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
    else:
        gray2 = img2

    sift = cv2.SIFT_create(nfeatures=nfeatures)

    kp1, des1 = sift.detectAndCompute(gray1, None)
    kp2, des2 = sift.detectAndCompute(gray2, None)

    matcher = cv2.FlannBasedMatcher()
    matches = matcher.knnMatch(des1, des2, k=2)

    good = []
    for m, n in matches:
        if m.distance < ratio_test * n.distance:
            good.append(m)

    src_pts = np.float32(
        [kp1[m.queryIdx].pt for m in good]
    )

    dst_pts = np.float32(
        [kp2[m.trainIdx].pt for m in good]
    )

    return src_pts, dst_pts
```

```python
def estimate_homography(src_pts, dst_pts,
                        ransac_thresh=5.0):
    """
    Estimate robust homography.

    Returns
    -------
    H : (3,3)
    inlier_mask : (N,)
    """

    H, mask = cv2.findHomography(
        src_pts,
        dst_pts,
        cv2.RANSAC,
        ransac_thresh
    )

    return H, mask.ravel().astype(bool)
```

```python
def warp_homography(img, H, output_shape):
    """
    Warp image with homography.

    Parameters
    ----------
    output_shape : (h, w)
    """

    h, w = output_shape

    return cv2.warpPerspective(
        img,
        H,
        (w, h),
        flags=cv2.INTER_LINEAR
    )
```

```python
def estimate_flow(img1, img2):
    """
    Dense Farneback flow from img1 -> img2.

    Returns
    -------
    flow : (H,W,2)
    """

    if img1.ndim == 3:
        gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    else:
        gray1 = img1

    if img2.ndim == 3:
        gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
    else:
        gray2 = img2

    flow = cv2.calcOpticalFlowFarneback(
        gray1,
        gray2,
        None,
        pyr_scale=0.5,
        levels=5,
        winsize=51,
        iterations=5,
        poly_n=7,
        poly_sigma=1.5,
        flags=0
    )

    return flow
```

```python
def warp_flow(img, flow):
    """
    Warp image using optical flow field.
    """

    h, w = flow.shape[:2]

    x, y = np.meshgrid(
        np.arange(w),
        np.arange(h)
    )

    map_x = (x + flow[..., 0]).astype(np.float32)
    map_y = (y + flow[..., 1]).astype(np.float32)

    return cv2.remap(
        img,
        map_x,
        map_y,
        cv2.INTER_LINEAR
    )
```

```python
import numpy as np


def hsl_to_rgb(hsl, output_uint8=False):
    """
    Convert HSL image to RGB.

    Parameters
    ----------
    hsl : ndarray (H, W, 3)
        H, S, L in range [0,1]

    output_uint8 : bool
        If True, returns uint8 [0,255].
        Otherwise returns float32 [0,1].

    Returns
    -------
    rgb : ndarray (H, W, 3)
    """
    h = hsl[..., 0]
    s = hsl[..., 1]
    l = hsl[..., 2]

    c = (1.0 - np.abs(2.0 * l - 1.0)) * s
    hp = h * 6.0
    x = c * (1.0 - np.abs((hp % 2.0) - 1.0))

    r1 = np.zeros_like(h)
    g1 = np.zeros_like(h)
    b1 = np.zeros_like(h)

    m0 = (0 <= hp) & (hp < 1)
    m1 = (1 <= hp) & (hp < 2)
    m2 = (2 <= hp) & (hp < 3)
    m3 = (3 <= hp) & (hp < 4)
    m4 = (4 <= hp) & (hp < 5)
    m5 = (5 <= hp) & (hp < 6)

    r1[m0], g1[m0] = c[m0], x[m0]
    r1[m1], g1[m1] = x[m1], c[m1]

    g1[m2], b1[m2] = c[m2], x[m2]
    g1[m3], b1[m3] = x[m3], c[m3]

    r1[m4], b1[m4] = x[m4], c[m4]
    r1[m5], b1[m5] = c[m5], x[m5]

    m = l - c / 2.0

    rgb = np.stack([
        r1 + m,
        g1 + m,
        b1 + m
    ], axis=-1)

    rgb = np.clip(rgb, 0.0, 1.0)

    if output_uint8:
        return (rgb * 255.0 + 0.5).astype(np.uint8)

    return rgb.astype(np.float32)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import rgb_to_hsv


def rgb_to_hsl(img):
    """
    Convert uint8 or float RGB image to HSL.

    Parameters
    ----------
    img : ndarray (H, W, 3)
        RGB image. uint8 [0,255] or float [0,1].

    Returns
    -------
    hsl : ndarray (H, W, 3)
        H, S, L in range [0,1]
    """
    img = img.astype(np.float32)

    if img.max() > 1.0:
        img /= 255.0

    r, g, b = img[..., 0], img[..., 1], img[..., 2]

    maxc = np.max(img, axis=-1)
    minc = np.min(img, axis=-1)

    l = (maxc + minc) / 2.0

    delta = maxc - minc

    s = np.zeros_like(l)

    mask = delta > 1e-8
    s[mask] = delta[mask] / (
        1.0 - np.abs(2.0 * l[mask] - 1.0)
    )

    h = np.zeros_like(l)

    rmask = mask & (maxc == r)
    gmask = mask & (maxc == g)
    bmask = mask & (maxc == b)

    h[rmask] = ((g[rmask] - b[rmask]) / delta[rmask]) % 6
    h[gmask] = ((b[gmask] - r[gmask]) / delta[gmask]) + 2
    h[bmask] = ((r[bmask] - g[bmask]) / delta[bmask]) + 4

    h /= 6.0

    return np.stack([h, s, l], axis=-1)


def plot_hsl_scatter(img1, img2, n_samples=20000, seed=42):
    """
    Scatter plot H,S,L values from img1 against img2.

    Parameters
    ----------
    img1, img2 : ndarray
        RGB images with same scene.
    n_samples : int
        Number of pixels to sample.
    seed : int
        RNG seed.
    """
    hsl1 = rgb_to_hsl(img1)
    hsl2 = rgb_to_hsl(img2)

    flat1 = hsl1.reshape(-1, 3)
    flat2 = hsl2.reshape(-1, 3)

    n = min(len(flat1), len(flat2))
    rng = np.random.default_rng(seed)

    idx = rng.choice(
        n,
        size=min(n_samples, n),
        replace=False
    )

    flat1 = flat1[idx]
    flat2 = flat2[idx]

    labels = ["Hue", "Saturation", "Lightness"]

    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    for i, ax in enumerate(axes):
        ax.scatter(
            flat1[:, i],
            flat2[:, i],
            s=1,
            alpha=0.2
        )

        ax.plot([0, 1], [0, 1], "r--", lw=1)

        ax.set_xlabel(f"{labels[i]} Image 1")
        ax.set_ylabel(f"{labels[i]} Image 2")
        ax.set_title(f"{labels[i]} Comparison")

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np

def apply_parametric_s_curve(image, alpha=3.0, beta=0.5, method='logistic'):
    """
    Applies a parametric S-curve to a post-gamma image normalized between [0, 1].
    
    Parameters:
    -----------
    image : ndarray
        Input image (NumPy array), values must be in the range [0.0, 1.0].
    alpha : float
        Contrast factor. 
        For 'logistic': alpha > 0 increases contrast (0 is identity).
        For 'rational': alpha > 1 increases contrast (1 is identity).
    beta : float
        The pivot/inflection point of the curve, typically between [0.1, 0.9].
        Values near 0.5 target midtones.
    method : str
        'logistic' for the industry standard, 'rational' for a faster algebraic curve.
        
    Returns:
    --------
    ndarray
        The contrast-enhanced image bounded perfectly between [0.0, 1.0].
    """
    # Safeguard input boundaries
    img_clipped = np.clip(image, 0.0, 1.0)
    
    if method == 'logistic':
        if alpha <= 0:
            return img_clipped
        
        # Internal helper for basic sigmoid
        def _sigmoid(x, a, b):
            return 1.0 / (1.0 + np.exp(-a * (x - b)))
        
        g_x = _sigmoid(img_clipped, alpha, beta)
        g_0 = _sigmoid(0.0, alpha, beta)
        g_1 = _sigmoid(1.0, alpha, beta)
        
        # Normalize boundaries to ensure perfect 0.0 to 1.0 mapping
        out = (g_x - g_0) / (g_1 - g_0)
        
    elif method == 'rational':
        if alpha <= 0:
            raise ValueError("Alpha must be greater than 0 for rational method.")
            
        # Avoid zero-division or log/power errors at exact 0.0 and 1.0 boundaries
        eps = 1e-7
        img_safe = np.clip(img_clipped, eps, 1.0 - eps)
        
        num = (img_safe ** alpha) * ((1.0 - beta) ** alpha)
        den = num + ((1.0 - img_safe) ** alpha) * (beta ** alpha)
        out = num / den
        
    else:
        raise ValueError("Method must be either 'logistic' or 'rational'.")
        
    return np.clip(out, 0.0, 1.0)
```

```python
import cv2
import numpy as np

def pixel_perfect_alignment(src_img, target_img, fine_tune_distortion=False):
    """
    Aligns src_img to target_img perfectly across the entire frame.
    Handles resolution mismatches, cropping, and non-linear distortion differences.
    
    Parameters:
        src_img (ndarray): The image to be warped/aligned.
        target_img (ndarray): The reference canvas template.
        fine_tune_distortion (bool): If True, uses dense optical flow to correct 
                                     pixel-level distortion discrepancies.
    """
    # --- STAGE 1: Prepare Grayscale Images ---
    gray_src = cv2.cvtColor(src_img, cv2.COLOR_BGR2GRAY)
    gray_target = cv2.cvtColor(target_img, cv2.COLOR_BGR2GRAY)
    
    # --- STAGE 2: Global Alignment (Scale, Translation, Crop) ---
    # Initialize SIFT detector for scale-invariant feature matching
    sift = cv2.SIFT_create()
    kp_src, des_src = sift.detectAndCompute(gray_src, None)
    kp_target, des_target = sift.detectAndCompute(gray_target, None)
    
    # Match features robustly
    bf = cv2.BFMatcher()
    matches = bf.knnMatch(des_src, des_target, k=2)
    
    # Filter matches using Lowe's ratio test
    good_matches = [m for m, n in matches if m.distance < 0.75 * n.distance]
    
    if len(good_matches) < 4:
        raise ValueError(f"Not enough matching features found ({len(good_matches)}/4 required).")
        
    # Extract coordinates of matched points
    src_pts = np.float32([kp_src[m.queryIdx].pt for m in good_matches]).reshape(-1, 1, 2)
    target_pts = np.float32([kp_target[m.trainIdx].pt for m in good_matches]).reshape(-1, 1, 2)
    
    # Compute Homography matrix with RANSAC to eliminate outlier matches
    H, _ = cv2.findHomography(src_pts, target_pts, cv2.RANSAC, 5.0)
    
    # Warp source to match the target's resolution scale and global positioning
    h_target, w_target = target_img.shape[:2]
    globally_aligned = cv2.warpPerspective(src_img, H, (w_target, h_target))
    
    if not fine_tune_distortion:
        return globally_aligned

    # --- STAGE 3: Local Alignment (Pixel-Perfect Distortion Fitting) ---
    # Convert globally aligned image to grayscale for dense flow analysis
    gray_globally_aligned = cv2.cvtColor(globally_aligned, cv2.COLOR_BGR2GRAY)
    
    # Calculate Dense Optical Flow (Farneback method) from target to aligned source.
    # This maps exactly how much each individual pixel shifted due to distortion.
    flow = cv2.calcOpticalFlowFarneback(
        gray_target, gray_globally_aligned, None,
        pyr_scale=0.5,    # Classical image pyramid scaling
        levels=5,         # Multi-scale pyramid depth
        winsize=21,       # Averaging window size (larger handles larger distortion changes)
        iterations=15,    # Iterations per pyramid level
        poly_n=7,         # Size of pixel neighborhood for polynomial expansion
        poly_sigma=1.5,   # Gaussian standard deviation for polynomial smoothing
        flags=0
    )
    
    # Generate pixel mapping grid
    map_x, map_y = np.meshgrid(np.arange(w_target), np.arange(h_target))
    map_x = map_x.astype(np.float32) + flow[..., 0]
    map_y = map_y.astype(np.float32) + flow[..., 1]
    
    # Remap image pixels using Bicubic interpolation for flawless sub-pixel accuracy
    pixel_perfect_aligned = cv2.remap(globally_aligned, map_x, map_y, cv2.INTER_CUBIC)
    
    return pixel_perfect_aligned

# Example Usage:
# target = cv2.imread("reference_corrected.jpg")
# source = cv2.imread("distorted_unaligned.jpg")
# result = pixel_perfect_alignment(source, target)
# cv2.imwrite("perfectly_aligned.jpg", result)
```
